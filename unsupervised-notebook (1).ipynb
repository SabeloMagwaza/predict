```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse
```

```python
# Load data
train_data = pd.read_csv("/kaggle/input/dataset-sabelo/train.csv")
test_data = pd.read_csv("/kaggle/input/dataset-sabelo/test.csv")
```

```python
# Data Preprocessing
reader = Reader(rating_scale=(1,5))
data = Dataset.load_from_df(train_data[['userId', 'movieId', 'rating']], reader)
```

```python
# Split data into training and testing sets
trainset, testset = train_test_split(data, test_size=0.01, random_state=42)
```

```python
# Model Selection: Singular Value Decomposition (SVD)
model = SVD()
```

```python
# Model Training
model.fit(trainset)
```

```python
# Model Evaluation on the testset
testset_predictions = model.test(testset)
rmse_score = rmse(testset_predictions)
print(f"Test RMSE: {rmse_score}")
```

```python
# Generate Predictions for submission
testset_to_predict = [(row['userId'], row['movieId'], 0) for _, row in test_data.iterrows()]
predictions = model.test(testset_to_predict)
```

```python
# Format predictions according to submission format
submission = pd.DataFrame([(f"{pred.uid}_{pred.iid}", pred.est) for pred in predictions], columns=['Id', 'rating'])
submission.to_csv("/kaggle/working/submission_predict.csv", index=False)
```
